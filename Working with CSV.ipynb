## 3. Opening a csv file from URL

## 14. Convertors

```python
## example we want short names of teams 
pd.read_csv('E:\\MyLearning\\datasets\\Working With CSV\\IPL Matches 2008-2020.csv')
```

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
pd.read_csv('E:\\MyLearning\\datasets\\Working With CSV\\IPL Matches 2008-2020.csv',converters={'team1':rename})
```
